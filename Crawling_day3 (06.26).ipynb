# 1페이지부터 10페이지까지 리뷰 수집

```python
from bs4 import BeautifulSoup as bs
import requests as req
```

```python
url = "https://movie.naver.com/movie/bi/mi/pointWriteFormList.nhn?code=189537&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false&page=1"
```

```python
for index in range(1,11,1):
    url = "https://movie.naver.com/movie/bi/mi/pointWriteFormList.nhn?code=189537&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false&page=" +str(index)
    res = req.get(url)
    info = res.text
    movie = bs(info, 'lxml')   #로딩할때마다 새로운페이지 계속하여 가공
    
    review = movie.select('div.score_reple > p')
    
    icon = movie.select('span.ico_viewer')
    spo = movie.select('span.text_spo._text_spo')
    
    
    for i in icon:
        i.extract()
    for j in spo:
        j.extract()
        
    for r in review:
        print(r.text.strip())
```

```python
for date in range(1,11,1):
    url = "https://movie.naver.com/movie/bi/mi/pointWriteFormList.nhn?code=189537&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false&page=" +str(date)
    res = req.get(url)
    movie = bs(res.text, 'lxml')
    
    review = movie.select('div.score_reple > p')
    
    icon = movie.select('span.ico_viewer')
    for i in icon:
        i.extract()
    spo = movie.select('span.text_spo._text_spo')
    for j in spo:
        j.extract()
        
    for r in review:
        print(r.text.strip())
```
