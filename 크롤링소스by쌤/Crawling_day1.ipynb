```python
## 가지고온 페이지 정보중에 내가 실제 필요한 데이터만 추출 할 때 사용
```

```python
from bs4 import BeautifulSoup as bs
import requests as req
```

```python
naver_url = "https://www.naver.com/"
```

```python
res = req.get(naver_url)
```

```python
info = res.text
```

```python
naver_res = bs(info,'lxml')
```

```python
list2 = naver_res.find_all('li',class_ = "nav_item")
```

```python
list2[0]
```

```python
list2[0].text
```

```python
list2[0].text.strip()
```

```python
for index in list2 : 
    print(index.text.strip())
```

#  네이버에서 코로나 검색후 뉴스제목만 크롤링

```python
from bs4 import BeautifulSoup as bs
import requests as req
```

```python
news_url = "https://search.naver.com/search.naver?sm=top_hty&fbm=0&ie=utf8&query=%EC%BD%94%EB%A1%9C%EB%82%98"
```

```python
rs = req.get(news_url)
```

```python
rs.text
```

```python
new_re = bs(rs.text,'lxml')
```

```python
news_list = new_re.find_all('a',class_= '_sp_each_title')
```

```python
news_list[0]
```

```python
news_list[0].text
```

```python
for index in news_list :
    print(index.text)
```
