# Beautiful Soup 사용하기

```python
## 가지고온 페이지 정보중에 내가 실제 필요한 데이터만 추출 할 때 사용
```

```python
from bs4 import BeautifulSoup as bs
import requests as req
```

```python
naver_url = "https://www.naver.com/"
```

```python
res = req.get(naver_url)        #잘 넘어왔는지 확인한 후 변수에 넣어줄 것
```

```python
info = res.text
```

```python
naver_res = bs(info, 'lxml')             #(가지고온 페이지의 내용, parsing방법(통상lxml사용))
```

```python
list2 = naver_res.find_all('li', class_="nav_item")           #(찾고자 하는  tag, 내가 구분지을것)       #ID만 찾아올때는 find

# class_ <<는 변하면 안됨. 변수X 명령어O
```

```python
list2[0]
```

```python
list2[0].text
```

```python
list2[0].text.strip()
```

```python
for index in list2:
    print(index.text.strip())
```

## 06.24  [장마 시작] 검색 후 뉴스제목 크롤링

```python
from bs4 import BeautifulSoup as bs
import requests as req
```

```python
rainy = "https://search.naver.com/search.naver?where=news&sm=tab_jum&query=%EC%9E%A5%EB%A7%88+%EC%8B%9C%EC%9E%91"
```

## [멕시코서 규모 7.4 강진] 검색 후 블로그제목 크롤링

```python
from bs4 import BeautifulSoup as bs
import requests as req
```
