```python
soup = bs(driver.page_source, 'lxml')
```

```python
name = soup.select('div#mCSB_3_container li.quickResultLstCon > strong')
```

```python
name[0].text.strip()
```

```python
addr = soup.select('div#mCSB_3_container li.quickResultLstCon > p.result_details')
```

```python
addr[0].text[:-9]       #주소
```

```python
addr[0].text[-9:]       #번호
```

```python
#len(name)
```

```python
#len(addr)
```

```python
rank_l = []
name_l = []
addr_l = []
tel_l = []

for i in range(len(name)):
    Rank = i + 1
    Name = name[i].text.strip()
    Addr = addr[i].text[:-9]
    Tel = addr[i].text[-9:]
    
    rank_l.append(Rank)
    name_l.append(Name)
    addr_l.append(Addr)
    tel_l.append(Tel)
    
print(rank_l)
print(name_l)
print(addr_l)
print(tel_l)
```

```python
sb_info = {'rank':rank_l, 'name':name_l, 'address':addr_l, 'tel':tel_l}
```

```python
sb_pd = pd.DataFrame(sb_info)
```

```python
sb_pd.set_index('rank', inplace = True)
```

# 이미지 저장 크롤링

```python
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup as bs
import time

import os       #폴더를 생성하거나 삭제할 때 사용

from urllib.request import urlretrieve  #(as aa처럼 써도 됨)      # 이미지 경로를  파일로 저장
```

```python
url = "https://search.naver.com/search.naver?sm=tab_hty.top&where=image&query=%EC%8B%9C%EA%B3%A0%EB%A5%B4%EC%9E%90%EB%B8%8C%EC%A2%85&oquery=%EA%BC%AC%EB%98%A5+%EB%93%9C+%ED%88%B4%EB%A0%88%EC%95%84&tqi=UYqCQdp0YidssjR9pGGssssssXh-274017"
```

```python
if not os.path.isdir('./이미지2') :    #is directory? = isdir    #os한테.경로찾으라~!. 이미지라는 폴더 존재하는지? (./는 찾아가는 것)
    print("폴더생성")
    os.mkdir('./이미지2')         #make directory = mkdir (위에 폴더 없으면 이미지라는 폴더 만들어라)
```

```python
driver = wb.Chrome()

driver.get(url)
```

```python
soup = bs(driver.page_source, 'lxml')
```

```python
img = soup.select('img._img')
```

```python
img[0]['src']          #text가 아니기때문에 .text하면 '' 나옴.    #['class']치면 클래스명 나옴
```

```python
img_list = []
for i in img :
    img_list.append(i['src'])
```

```python
img_list
```

```python
len(img_list)
```

```python
import time
```

```python
for index in range(len(img_list)):
    urlretrieve(img_list[index],'./이미지/' +str(index) +'.jpg')
    
    time.sleep(2)
    
#urlretrieve 아까가져온거룰 변환해주고, 
#폴더에 넣을거야 ((~이미지까지는 폴더고 /부터는 파일))
## 무조건 스크롤하고 다운받아야함(이미지 로딩되야 하기 때문)
```
